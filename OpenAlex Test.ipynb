Define filters we want to search on: 
- abstract (should include physics?)
- type should be article
- countries/continent?
- year?
- institution?


```python
pages=[]

pager = Works().search_filter(abstract="physics") \
        .filter(type='article') \
        .filter(authorships={"institutions": {"continent": "europe"}}) \
        .filter(publication_year="2020") \
        .paginate(per_page=200)

for page in pager:
    pages.append(len(page))
    print(len(page))


print('Total length:', sum(pages))
```

```python
physics_data=[]

query = Works().search_filter(abstract="physics") \
        .filter(type='article') \
        .filter(authorships={"institutions": {"continent": "europe"}}) \
        .filter(publication_year="2020")

for record in chain(*query.paginate(n_max=None)):
    physics_data.append(record)
```
