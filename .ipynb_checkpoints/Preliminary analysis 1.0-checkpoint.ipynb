## Import data

```python
%%time

physics_data=[]

query = Works().search_filter(abstract="physics") \
        .filter(type='article') \
        .filter(authorships={"institutions": {"continent": "europe"}}) \
        .filter(publication_year="2023")

for record in chain(*query.paginate(n_max=None)):
    physics_data.append(record)
```
